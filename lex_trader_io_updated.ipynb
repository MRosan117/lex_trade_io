```python
# Imports
import pandas as pd
import numpy as np
import pandas_ta as ta
from pathlib import Path
import hvplot.pandas
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.preprocessing import StandardScaler
from pandas.tseries.offsets import DateOffset
from sklearn.metrics import classification_report
from finta import TA

import warnings
warnings.filterwarnings('ignore')
```

```python
# Filter the date index and close columns
# combined_clean_data_close = combined_clean_data.loc[:, ["Close"]]

# Use the pct_change function to generate returns from close prices
clean_spy_data["Actual Returns"] = clean_spy_data["Close"].pct_change()

# Drop all NaN values from the DataFrame
clean_spy_data = clean_spy_data.dropna()

# Review the DataFrame
display(clean_spy_data.head())
# display(combined_clean_data.tail())
```

```python
clean_spy_data['RSI'] = clean_spy_data.ta.rsi(length=14)

def myRSI(price, n=20):
    delta = price['close'].diff()
    dUp, dDown = delta.copy(), delta.copy()
    dUp[dUp < 0] = 0
    dDown[dDown > 0] = 0

    RolUp = dUp.rolling(window=n).mean()
    RolDown = dDown.rolling(window=n).mean().abs()

    RS = RolUp / RolDown
    rsi = 100.0 - (100.0 / (1.0 + RS))

clean_spy_data.head(20)
```

```python
dfpl = clean_spy_data
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from datetime import datetime

fig = make_subplots(rows=2, cols=1)
fig.append_trace(go.Candlestick(x=dfpl.index,
                                open=dfpl['Open'],
                                high=dfpl['High'],
                                low=dfpl['Low'],
                                close=dfpl['Close']), row=1, col=1)
fig.append_trace(go.Scatter(
    x=dfpl.index,
    y=dfpl['RSI'],
), row=2, col=1)

fig.update_layout(
    margin=dict(l=40, r=40, t=10, b=0),
    paper_bgcolor="LightSteelBlue",
    autosize=False,
    width=1000,
    height=1000,
 )

fig.update_layout(xaxis_rangeslider_visible=False)
fig.show()
```

```python
# Initialize the new Signal column
clean_spy_data['RSI Signal'] = 0.0

# When Actual Returns are greater than or equal to 0, generate signal to buy stock long
clean_spy_data.loc[(clean_spy_data['RSI'] >= 52), 'RSI Signal'] = 1

# When Actual Returns are less than 0, generate signal to sell stock short
clean_spy_data.loc[(clean_spy_data['RSI'] < 48), 'RSI Signal'] = -1

# Review the DataFrame
display(clean_spy_data.head(25))
display(clean_spy_data.tail())
```

```python
clean_spy_data['RSI Signal'].value_counts()
```

```python
clean_spy_data['Actual Returns'] = clean_spy_data['Close'].pct_change()
```

```python
clean_spy_data.tail(50)
```

```python
# Calculate the strategy returns and add them to the signals_df DataFrame
clean_spy_data['RSI Strategy Returns'] = clean_spy_data['Actual Returns'] * clean_spy_data['RSI Signal'].shift()
```

```python
display(clean_spy_data.head())
display(clean_spy_data.tail())
```

```python
# Generate the X_train and y_train DataFrames
X_train = X.loc[training_begin:training_end]
y_train = y.loc[training_begin:training_end]

# Review the X_train DataFrame
X_train.head()
```

```python
display(training_end)
```

```python
# Generate the X_test and y_test DataFrames
X_test = X.loc[training_end:]
y_test = y.loc[training_end:]

# Review the X_test DataFrame
display(X_train.head())
display(X_test.tail())
```

```python
display(X_test)
```

```python
# Scale the features DataFrames

# Create a StandardScaler instance
scaler = StandardScaler()

# Apply the scaler model to fit the X-train data
scaler.fit(X_train)


# Transform the X_train and X_test DataFrames using the X_scaler
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# From SVM, instantiate SVC classifier model instance
svm_model = svm.SVC()

# Fit the model to the data using the training data
svm_model = svm_model.fit(X_train_scaled, y_train)

# Use the testing data to make the model predictions
svm_pred = svm_model.predict(X_test_scaled)

# Review the model's predicted values
svm_pred[:10]
```

```python
# Use a classification report to evaluate the model using the predictions and testing data
svm_testing_report = classification_report(y_test, svm_pred)

# Print the classification report
print(svm_testing_report)
```

```python
# Create a new empty predictions DataFrame:

# Create a predictions DataFrame
predictions_df = pd.DataFrame(index=X_test.index)

# Add the SVM model predictions to the DataFrame
predictions_df['Predicted'] = svm_pred

# Add the actual returns to the DataFrame
predictions_df['Actual Returns'] = clean_spy_data["Actual Returns"]

# Add the strategy returns to the DataFrame
predictions_df['Strategy Returns'] = (
    predictions_df["Actual Returns"] * predictions_df["Predicted"]
)

# Review the DataFrame
display(predictions_df.head())
display(predictions_df.tail())
```

```python
# Plot the actual returns versus the strategy returns
(1 + predictions_df[["Actual Returns", "Strategy Returns"]]).cumprod().plot()
```

```python
annualized_return = predictions_df["Strategy Returns"].mean() * 252
annualized_std = predictions_df["Strategy Returns"].std() * np.sqrt(252)
sharpe_ratio = round(annualized_return/annualized_std, 3)

# Display results
print("Sharpe Ratio:", sharpe_ratio)
```

```python
#Best-Case/Worst-Case Performance
predictions_df["Strategy Returns"].plot()
```

```python
# Create a trading algorithm using kama
# Set the Signal column
clean_spy_data["MACD Signal"] = 0.0

# Generate the trading signals 1 (entry) or -1 (exit) for a long position trading algorithm
# where 1 is when the Close price is less than the kama window
# where -1 is when the Close price is greater the the close
for index, row in clean_spy_data.iterrows():
    if row["Close"] < row["MACD"]:
        clean_spy_data.loc[index, "MACD Signal"] = 1.0
    if row["Close"] > row["MACD"]:
        clean_spy_data.loc[index,"MACD Signal"] = -1.0

# Review the DataFrame
clean_spy_data
```
